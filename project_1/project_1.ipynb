```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from statsmodels.tsa.stattools import adfuller
```

```python
df_global = pd.read_csv("results-7.csv")
```

```python
df_global.head()
```

```python
# #of null values in the global data set
print(df_global.isnull().sum().sum())
```

```python
x_bar_global = df_global.avg_temp.mean()
var_global = df_global.avg_temp.var()
print("The mean global temperature is: ",x_bar_global)
print("The global variance is: ",var_global)
print("The max global temperature is: ",df_global.avg_temp.max())
print("The max global temperature is: ",df_global.avg_temp.min())
```

```python
x_bar_series_global = np.full(shape=df_global.shape[0], fill_value=x_bar_global, dtype=np.float64)
x_bar_series_global = pd.Series(x_bar_series_global)
plt.plot(df_global.year,df_global.avg_temp)
plt.plot(df_global.year,x_bar_series_global)
plt.xlabel('Time (years)')
plt.ylabel('Average Temperature')
plt.title('Global Temperature vs. Time')
plt.legend(["Global Temperature","Global Mean"])
```

```python
df_city = pd.read_csv("results-8.csv")
x_bar_city = df_city.avg_temp.mean()
var_city = df_city.avg_temp.var()
print("The mean temperature in Atlanta is: ",x_bar_city)
print("The varince in the temperature in Atlanta is: ",var_city)
print("The min temperature in Atlanta is: ",df_city.avg_temp.min())
print("The max temperature in Atlanta is: ",df_city.avg_temp.max())
print(var_city)
```

```python
x_bar_series_city = np.full(shape=df_city.shape[0], fill_value=x_bar_city, dtype=np.float64)
x_bar_series_city = pd.Series(x_bar_series_city)
print(df_city.shape[0])
```

```python
plt.plot(df_city.year,df_city.avg_temp)
plt.plot(df_city.year,x_bar_series_city)
plt.xlabel('Time (years)')
plt.ylabel('Temperature')
plt.title('Atlanta Temperature vs. Time')
plt.legend(["Moving Average","Mean Value"])
```

```python
global_rolling_mean = df_global.avg_temp.rolling(window=5).mean()
plt.plot(df_global.year,global_rolling_mean)
plt.plot(df_global.year,x_bar_series_global)
plt.xlabel('Time (years)')
plt.ylabel('Global Average Temperature (C)')
plt.title('Global Moving Average Temperature vs. Time')
plt.legend(["Global Moving Average","Global Mean"])
```

```python
city_rolling_mean = df_city.avg_temp.rolling(window=5).mean()
plt.plot(df_city.year,city_rolling_mean)
plt.plot(df_city.year,x_bar_series_city)
plt.xlabel('Time (years)')
plt.ylabel('Atlanta Average Temperature (C)')
plt.title('Atlanta Moving Average Temperature vs. Time')
plt.legend(["Atlanta Moving Average","Atlanta Mean"])
#city_rolling_mean.head(50)
```

```python
plt.plot(df_city.year,city_rolling_mean)
plt.plot(df_city.year,df_city.avg_temp)
plt.plot(df_city.year,x_bar_series_city)
plt.legend(["Moving Average","Atlanta Average Temperature","Atlanta Mean"])
plt.title("Atlanta Temperature vs. Time")
plt.xlabel("Time (years)")
plt.ylabel("Temperature")
plt.savefig("atlantaTemp.png")
```

```python
plt.plot(df_global.year,global_rolling_mean,c = 'purple')
plt.plot(df_global.year,df_global.avg_temp,c = 'green')
plt.plot(df_global.year,x_bar_series_global,c='orange')
plt.legend(["Moving Average","Global Average Temperature"])
plt.title("Global Temperature vs. Time")
plt.xlabel("Time (years)")
plt.ylabel("Temperature")
plt.savefig("globalTemp.png")
```

```python
df_city_sliced = pd.concat([df_city['year'],df_city['avg_temp']],axis=1)
df_city_sliced.columns = ['year_city','avg_temp_city']
city_rolling_mean.columns = ['city_moving_average']
global_rolling_mean.columns = ['global_moving_average']
global_rolling_mean.head()
city_rolling_mean.head()
```

```python
my_data_frames = [df_global,df_city_sliced,city_rolling_mean,global_rolling_mean,x_bar_series_city,
                 x_bar_series_global]
combined_data_frame = pd.concat(my_data_frames,sort = False,axis=1)
combined_data_frame = combined_data_frame.dropna(axis=0)
combined_data_frame.columns = ['year','global_temp','year_city','Atlanta_temp',
                               'city_moving_average','global_moving_average','mean_temp_atlanta','mean_temp_global']
combined_data_frame.head()
```

```python
plt.plot(combined_data_frame['year'],combined_data_frame['global_temp'])
plt.plot(combined_data_frame['year'],combined_data_frame['Atlanta_temp'])
plt.plot(combined_data_frame['year'],combined_data_frame['city_moving_average'])
plt.plot(combined_data_frame['year'],combined_data_frame['global_moving_average'])
plt.xlabel('Time (years)')
plt.ylabel('Temperature (Celsius)')
plt.title('Temperature vs Time')
plt.legend(['global_temp','Atlanta_temp','Atlanta_moving_avg','Global_moving_avg'])
plt.savefig('globalAndAtlantaRawData.png')
```

```python
plt.plot(combined_data_frame['year'],combined_data_frame['city_moving_average'])
plt.plot(combined_data_frame['year'],combined_data_frame['global_moving_average'])
plt.plot(combined_data_frame['year'],combined_data_frame['mean_temp_atlanta'])
plt.plot(combined_data_frame['year'],combined_data_frame['mean_temp_global'])
plt.title('Temperature vs. Time')
plt.xlabel('Time (years)')
plt.ylabel('Temperature (Celsius)')
plt.legend(['Atlanta_moving_average','gobal_moving_average','mean_temp_atlanta',
           'mean_temp_atlanta'])
plt.savefig('globalAndAtlanta.png')
```

```python
def adf_test(ts):
    print('Dickey-Fuller Test Results:')
    df_test = adfuller(ts,autolag = 'AIC')
    df_out = pd.Series(df_test[0:4],index = ['Test Statistic','p-value','Lags Used',
                                           'Number of Observations Used'])
    for key,value in df_test[4].items():
        df_out['Critical Value (%s)'%key] = value
    print(df_out)
```

```python
adf_test(combined_data_frame['global_temp'])
```

```python
adf_test(combined_data_frame['Atlanta_temp'])
```
