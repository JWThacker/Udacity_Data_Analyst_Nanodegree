```python
import requests
```

```python
r = requests.get('https://imgs.xkcd.com/comics/python.png')
```

```python
#Prints attributes and methods associated with the above response object
print(dir(r))
```

```python
#Prints the .html acquired from the above URL.
print(r.text)
```

```python
#Prints helpful information related to attributes and methods associated with the response object.
print(help(r))
```

```python
with open('comic.png','wb') as f:
    f.write(r.content)
```
