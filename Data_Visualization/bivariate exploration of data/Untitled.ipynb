```python
def log_trans(x,inverse=False):
    if not inverse:
        return np.log10(x)
    else:
        return np.power(10,x)

sb.regplot(fuel_econ['displ'],fuel_econ['comb'].apply(log_trans));
tick_locs = [10,20,50,100]
plt.yticks(log_trans(tick_locs),tick_locs);
```

```python
displ_desc = fuel_econ.displ.describe()
comb_desc = fuel_econ.comb.describe()
displ_min = displ_desc.loc['min']
displ_max = displ_desc.loc['max']
comb_min = comb_desc.loc['min']
comb_max = comb_desc.loc['max']
```

```python
max_min_arr = [displ_min,displ_max,comb_min,comb_max]
for i in max_min_arr:
    print(i)
```

```python
bins_x = np.arange(displ_min - 0.5,displ_max+0.3,0.3)
bins_y = np.arange(comb_min-0.5,comb_max+3,3)
#bins_x

plt.hist2d(data=fuel_econ,x='displ',y='comb',cmin=0.5,cmap='plasma',bins=[bins_x,bins_y]);
plt.colorbar()
plt.xlabel('Displacement (1)');
plt.ylabel('Combined Fuel Eff. (mpg)');
```

```python
bins_x = np.arange(0.6, 7+0.7, 0.7)
bins_y = np.arange(12, 58+7, 7)
h2d = plt.hist2d(data=fuel_econ,x='displ',y='comb',cmin=0.5,cmap='plasma',bins=[bins_x,bins_y])
plt.colorbar();
plt.xlabel('Displacement (l)');
plt.ylabel('Combined Fuel Eff. (mpg)')
counts = h2d[0]

for i in range(counts.shape[0]):
    for j in range(counts.shape[1]):
        c = counts[i,j]
        if c>= 100:
            plt.text(x=bins_x[i]+0.3,y=bins_y[j]+1.0,s=int(c),ha='center',va='center',color='black')
        elif c > 0:
            plt.text(x=bins_x[i]+0.3,y=bins_y[j]+1.0,s=int(c),ha='center',va='center',color='white')
```
